```python
import pyfolio as pf
import yfinance as yf
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

# 1. Define the assets and how much money goes into each (weights)
tickers = ['NVDA', 'GS', 'MSFT']
weights = pd.Series([0.40, 0.40, 0.20], index=tickers)

# 2. Grab closing price histories for all three assets
prices = pd.DataFrame()
for ticker in tickers:
    stock_data = yf.Ticker(ticker).history(period="max")['Close']
    prices[ticker] = stock_data

# Clean up matching dates and convert to UTC timezone
prices = prices.dropna()
prices.index = pd.to_datetime(prices.index).tz_convert('UTC')

# 3. Calculate separate asset returns and combine them using our weights
asset_returns = prices.pct_change().dropna()
portfolio_returns = asset_returns.multiply(weights, axis=1).sum(axis=1)

# 4. Fetch the general S&P 500 Market Benchmark
bench = yf.Ticker("SPY").history(period="max")
bench_returns = bench['Close'].pct_change().dropna()
bench_returns.index = pd.to_datetime(bench_returns.index).tz_convert('UTC')

# 5. Align timelines
portfolio_returns, bench_returns = portfolio_returns.align(bench_returns, join='inner')

# 6. Set our evaluation live split date
clean_live_date = pd.Timestamp('2023-01-01').tz_localize('UTC')

# 7. Run the combined portfolio analysis!
pf.create_returns_tear_sheet(portfolio_returns, live_start_date=clean_live_date, benchmark_rets=bench_returns)
```
